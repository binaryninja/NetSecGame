```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
models = ["Random", "RandomHeur", "Q-Learning", "LLM", "LLM ReAct"] # "Double Q",
max_steps = [30, 60, 100]
```

## No Defender

### Return

```python
random_mean = np.array([-26.041, -37.828, -43.578])
random_std = np.array([20.220, 45.726, 69.071])

random_heur_mean = np.array([-3.775, 7.6, 19.244])
random_heur_std = np.array([45.914, 62.893, 65.705])

q_agent_mean = np.array([51.16, 57.26, 47.19])

llm_qa_mean = np.array([-6.633, 76.182, 80])
llm_qa_std = np.array([46.759, 13.354, 10])

llm_qa_4_mean = np.array([70.233, 83.100, 83.100])
llm_qa_4_std = np.array([39.746, 8.761, 8.761])
```

```python
(65.475 + 23.083 +53.041)/3
```

```python
plt.plot(max_steps, random_mean, marker=".", label="Random")
# plt.fill_between(max_steps, random_mean-random_std, random_mean+random_std, alpha=0.2)

plt.plot(max_steps, random_heur_mean, marker="d", label="Random Heur")
# plt.fill_between(max_steps, random_heur_mean-random_heur_std, random_heur_mean+random_heur_std, alpha=0.2)

plt.plot(max_steps, q_agent_mean, marker="d", label="Q-Agent")
# plt.fill_between(max_steps, q_agent_mean-q_agent_mean, q_agent_mean+q_agent_mean, alpha=0.2)

plt.plot(max_steps, llm_qa_mean, marker="*", label="LLM ReAct(GPT-3.5)")
# plt.fill_between(max_steps, llm_qa_mean-llm_qa_std, llm_qa_mean+llm_qa_std, alpha=0.2)

plt.plot(max_steps, llm_qa_4_mean, marker="^", label="LLM ReAct(GPT-4)")
# plt.fill_between(max_steps, llm_qa_4_mean-llm_qa_4_std, llm_qa_4_mean+llm_qa_4_std, alpha=0.2)

plt.grid(axis='y')
plt.ylabel("Average Return")
plt.xlabel("Max Steps")
plt.xticks(max_steps)
plt.ylim(-60, 100)
plt.legend()
# plt.title("Average Return - No Defender")
```
