```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
models = ["Random", "RandomHeur", "LLM", "LLM ReAct"] # "Q-Learning", "Double Q",
max_steps = [30, 60, 100]
```

## No Defender

### Return

```python
random_mean = [-26.041, -37.828, -43.578]
random_std = [20.220, 45.726, 69.071]

random_heur_mean = [-3.775, 7.6, 19.244]
random_heur_std = [45.914, 62.893, 65.705]

llm_3_mean = [-6.633, 76.182, 80]
llm_3_std = [46.759, 13.354, 10]
```

```python
plt.plot(max_steps, random_mean, marker=".", label="Random")
plt.plot(max_steps, random_heur_mean, marker="d", label="Random Heur")
plt.plot(max_steps, llm_3_mean, marker="*", label="LLM 3(GPT3.5)")

plt.grid(axis='y')
plt.ylabel("Average Return")
plt.xlabel("Max Steps")
plt.xticks(max_steps)
plt.ylim(-60, 100)
plt.legend()
plt.title("Average Return - No Defender")
```

### Win Rate

```python
random_mean = [3.7, 18.35, 41.1]
random_heur_mean = [24.65, 54.15, 78.55]
# llm_3_mean = [-6.633, 76.182, 80]
```

```python
plt.plot(max_steps, random_mean, marker=".", label="Random")
plt.plot(max_steps, random_heur_mean, marker="d", label="Random Heur")
# plt.plot(max_steps, llm_3_mean, marker="*", label="LLM 3(GPT3.5)")

plt.grid(axis='y')
plt.ylabel("Win Rate (%)")
plt.xlabel("Max Steps")
plt.xticks(max_steps)
plt.ylim(0, 100)
plt.legend()
plt.title("Win Rate - Stochastic w Threshold Defender")
```
