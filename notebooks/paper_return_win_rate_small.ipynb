```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('default')
plt.style.use('fivethirtyeight')
plt.style.use('seaborn-v0_8-white')
```

```python
models = ["Random", "RandomHeur", "Q-Learning", "LLM", "LLM ReAct"]
max_steps = [30, 60, 100]
```

## No Defender

### Return

```python
random_mean = [-26.246, -37.178, -43.433]
random_heur_mean = [-5.639, 8.470, 16.289]
q_agent_mean = [43.93, 47.547, 16.69]
llm_mean = [4.367, 4.367, -16.133]
llm_qa_mean = [70.233, 83.100, 83.100]
```

```python
plt.rcParams.update({'font.size': 20})

x = np.arange(len(max_steps))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(16, 8))
rects2 = ax.bar(x - width/2, random_mean, width/4, capsize=5, label='Random')
rects2 = ax.bar(x - width/4, random_heur_mean, width/4, capsize=5, label='Random Heur')
rects3 = ax.bar(x, q_agent_mean, width/4, capsize=5, label='Q Learning')
rects4 = ax.bar(x + width/4, llm_mean, width/4, label='LLM (One-shot)')
rects5 = ax.bar(x + width/2, llm_qa_mean, width/4, capsize=5, label='LLM (ReAct)')
ax.set_ylabel('Average Return')
ax.set_xlabel('Max Steps')
ax.set_xticks(x)
ax.set_xticklabels(max_steps)
# ax.set_yticks(range(10, 110, 10))


ax.legend(loc=(1.02, 0.3))
ax.yaxis.grid(True, linestyle='-.')

fig.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

plt.plot(max_steps, random_mean, marker=".", label="Random")
plt.plot(max_steps, random_heur_mean, marker="d", label="Random Heur")
plt.plot(max_steps, q_agent_mean, marker="o", label="Q-Agent")
plt.plot(max_steps, llm_mean, marker="*", label="LLM (One-shot)")
plt.plot(max_steps, llm_qa_mean, marker="^", label="LLM (ReAct)")

plt.grid(axis='y')
plt.ylabel("Average Return")
plt.xlabel("Max Steps")
plt.xticks(max_steps)
plt.ylim(-100, 100)
plt.legend()

ax.legend(loc=(1.02, 0.3))
ax.yaxis.grid(True, linestyle='-.')
# plt.title("Average Return - No Defender")
```
